```python
from scipy.optimize import curve_fit
results = []
for ts in test:
    x = np.arange(len(ts))
    [a,b,c], res = curve_fit(lambda t,a,b,c: a*((t+c)**2) + b,  x,  ts)
    print(a,b,c)
    results.append([a,b,c])
```
